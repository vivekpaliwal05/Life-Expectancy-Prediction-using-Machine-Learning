```python
x=df_copy[['Age', 'age_score', 'Gender', 'Systolic_bp', 'systolic_score', 'Systolic_Category' , 'Distolic_bp', 'distolic_scorew',  'cholesterol_score', 'Diabetes', 'Diabetes_score ', 'Heart', 'heart_score', 'Smoking', 'smoking_score', 'Alcohol', 'alchol_score', 'Physical_Activity', 'physical_score', 'Sleep_Hours', 'BMI_Calculated', 'bmi_score' ]] 
y=df_copy['risk_score']
```

```python
x.isnull()
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2 ,random_state=42)
```

```python
x_scaled =StandardScaler()
x_train = x_scaled.fit_transform(x_train)
x_test=x_scaled.transform(x_test)
```

```python
model = LinearRegression()
model.fit(x_train , y_train)
y_predict = model.predict(x_test)
```

```python
print(y_predict)
```

```python
print("R2_SCORE  : " , r2_score(y_test ,y_predict))
print("MEAN_ABSOLUTE_ERROR : " , mean_absolute_error(y_test , y_predict))
print("MEAN _SQUARED_EROR : " , mean_squared_error(y_test ,y_predict))
```
